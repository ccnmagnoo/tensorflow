````python
import tensorflow as tf
import numpy as np
import pandas as pd
````

Data will be the base of data Model

````python
celcius = np.array([-40,-10,0,8,15,22,38],dtype=float)
fahrengeit=np.array([-40,14,32,46,59,72,100],dtype=float)
````

Creating layers
```
Dense() function 
#set interaction with Many to Many neurones
```



## configurando la forma del modelo
Con **una** neurona de salida y **una** entrada

````python
capa = tf.keras.layers.Dense(units=1,input_shape=[1])
modelo = tf.keras.Sequential([capa])
````

````python
modelo.compile(
    optimizer=tf.keras.optimizers.Adam(0.1),
    loss='mean_squared_error'
)
````

````python
print('init model')
#instanciando modelo
historial = modelo.fit(celcius,fahrengeit,epochs=1000,verbose=False)
print('finalice model')
````

````python
import matplotlib.pyplot as plt
plt.xlabel('# Epoca')
plt.ylabel('Magnitud Perdida')
plt.plot(historial.history['loss'])
````

## Predicción

````python
print('Predicción')
resultado = modelo.predict([100.0,0])
print('result of prediction ' +str(resultado) +' °F' )
````

````python
print('Variables internas del modelo')
print(capa.get_weights())
````

##Con varias neuronas
con 3 neuronas, 2 ocultas + 1 salida, y 1 entrada

````python
#capas
oculta1 = tf.keras.layers.Dense(units=2,input_shape=[1])
oculta2 = tf.keras.layers.Dense(units=2)
salida = tf.keras.layers.Dense(units=1)
modelo2 = tf.keras.Sequential([oculta1,oculta2,salida])
````

````python
#compilando modelo
modelo2.compile(
    optimizer=tf.keras.optimizers.Adam(0.1),
    loss='mean_squared_error'
)
````

````python
#entrenando modelo
print('iniciando modelo')
historial2 = modelo2.fit(celcius,fahrengeit,epochs=1000,verbose=False)
print('modelo más complejo entrenado')
````

````python
plt.plot(historial2.history['loss'])
````

````python
print('Variables internas del modelo')
import pprint
pp = pprint.PrettyPrinter(indent=4)
````

````python
pp.pprint(oculta1.get_weights())
pp.pprint(oculta2.get_weights())
pp.pprint(salida.get_weights())
````
